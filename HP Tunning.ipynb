Imports Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
np.random.seed(1)
```

```python
boston = np.genfromtxt("boston.csv", delimiter=",", skip_header=1)
X, y = boston[:,:-1], boston[:,-1]
```

Split Dataset

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.8)
```

Define Model, Train, Test and display the results

```python
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge(alpha=1)
ridge.fit(X_train, y_train)

y_pred = ridge.predict(X_test)
mean_squared_error(y_test, y_pred)
```

Do grid search for alpha

```python
result = []
for a in np.linspace(-0.5, 0.5, 10):
    ridge = Ridge(alpha = a)
    ridge.fit(X_train, y_train)

    y_pred = ridge.predict(X_test)
    result += [[a, mean_squared_error(y_test, y_pred)]]
```

Plot the grid search

```python
result = np.array(result)
plt.plot(result[:, 0], result[:, 1])
plt.show()
```
